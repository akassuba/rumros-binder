### 2. Launch RuMROS with either ARGoS (recommended) or Gazebo (legacy mode):

```python
import subprocess
from time import sleep
import ipywidgets as widgets

# Start mosquitto
subprocess.Popen(["gnome-terminal", "--", "bash", "-c", "mosquitto -d"])

# Create a dropdown
launcher = widgets.Dropdown(
    options=[
        ("ARGOS", "argos"),
        ("Gazebo", "gazebo")
    ],
    value="argos",
    description="Simulator:"
)

display(launcher)

# Launch button
button = widgets.Button(description="Launch")

display(button)

def launch(_):
    if launcher.value == "argos":
        script = """
        cd /app/rumros
        chmod +x launch_with_argos.sh
        ./launch_with_argos.sh
        """
    else:
        script = """
        cd /app/rumros
        chmod +x launch_with_gazebo.sh
        ./launch_with_gazebo.sh
        """

    subprocess.Popen(["gnome-terminal", "--", "bash", "-c", script])

    print("Starting...")
    sleep(15)

button.on_click(launch)
```

### 5.  Play around with RuMROS

The dashboard displayed on the web interface shows tables with information available to the runtime model, including group structure, current robot positions and more. The sidebar on the right allows actions to be executed on groups and individual robots. For Crazyflies, Takeoff, Land, Go to position and Go to position (relative) are supported, the other behaviors are intended for ground robots. Simply choose an appropriate action, target group/unit and launch it. The second and third pages in the navigation bar on top display tables and actions on a larger page for a better overview, respectively.

On the fourth page, the visualization of the behavior model implemented in 'ArgosMainLoop.jadd' resides. It shows the HFSM structure, with composite nodes (CBs), leaf nodes (BBs) and their associated behaviors and transitions. The example warehouse scenario is fairly complex, so you may have to move nodes around by dragging them in order to get a better overview of the structure. CBs have a name that indicates what behavior they perform, while contained BBs have an ID, last start and end times, as well as name of their associated behavior. Transitions are either annotated with the time after which they transition, a timestamp at which they transition, or are dashed and without text. Refer to the legend for more details on the modeling notation. To start the behavior model, press the start button on the bottom right, and the orange starting behaviors should turn green, indicating the model was launched. Make sure to press the play button beforehand in the simulator (ARGoS), and the robots should start moving accordingly.
